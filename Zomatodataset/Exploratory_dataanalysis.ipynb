```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head(2)
```

```python
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values
```

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

##### Observations
the presence of zomato maybe in India as highest transactions are happening here

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

#### Observations
1. when rating is between 4.5 to 4.9 ---> Excellent
2. when rating is between 4.0 to 4.4 ---> Very Good
3. when rating is between 3.5 to 3.9 ---> Good
4. when rating is between 2.5 to 3.4 ---> Average
5. when rating is between 1.8 to 2.4 ---> Poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```

```python
country_names_with_zero_rating=final_df[final_df['Aggregate rating']==0.0].groupby(['Country']).size()
print(country_names_with_zero_rating)
print(list(country_names_with_zero_rating.index))
```

#### Observation:
The __Maximum No of 0__ Ratings are from India
